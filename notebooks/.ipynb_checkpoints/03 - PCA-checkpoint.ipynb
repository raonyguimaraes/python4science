A Análise de Componentes Principais ou principal component analysis (PCA) é um procedimento matemático que utiliza uma transformação ortogonal para converter um conjunto de observações de variáveis possivelmente correlacionadas a um conjunto de valores de variáveis linearmente descorrelacionadas chamadas componentes principais.

```python
%matplotlib inline
from sklearn import decomposition

#PCA
from sklearn import cluster, datasets
iris = datasets.load_iris()

import pylab as pl

pca = decomposition.PCA(n_components=3)
pca.fit(iris.data)
decomposition.PCA(copy=True, n_components=3, whiten=False)
X = pca.transform(iris.data)

pl.scatter(X[:, 0], X[:, 1], c=iris.target)
```

```python
%matplotlib qt

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=iris.target)

plt.show()
```
